```python
# Data Frame de Atividade Física

import pandas as pd

data = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
atividade = ['Jiu-Jitsu', 'Musculação', 'Jiu-Jitsu', 'Musculação', 'Jiu-Jitsu', 'Descanso']
duração = [1.30, 1, 1.30, 1, 1.30, 0]
especificação = ['Treino Pesado', 'Treino Pesado', 'Treino Técnico', 'Treino Pesado', 'Treino Leve', 'Descanso']
calorias_queimadas = [220, 250, 200, 220, 180, 0]
AtividadeFisica = {"Data": data, "Atividade": atividade, "Duração": duração, "Especificação": especificação, "Calorias Queimadas": calorias_queimadas}

df = pd.DataFrame(AtividadeFisica)

display(df)



```

# Questão 07

```python
# Você deve adicionar um novo item (linha) à base de dados
print("Adicionar uma nova linha\n")

item = ['Domingo', 'Descanso', 0, 'Descanso', 0] # Itens a serem adicionados

df.loc[len(df)] = item # Localizando o tamanho do meu data frame e adicionando ao final os novos itens
print(f"Foram incluidos os dados {item} na linha {len(df)-1}  do seu DataFrame\n")
print(df)
```

# Questão 08

```python
# Você deve realizar a transposição de uma coluna para uma linha na base de dados
print("Transposição\n")

df_t = df.T # Transformando linha em coluna
display(df_t) # Mostrando o data frame
print("O Data Frame foi transposto")

linhas, coluna = df_t.shape
print(f"O Data Frame apresenta {coluna} colunas")
print(f"O Data Frame apresenta {linhas} linhas")
```
